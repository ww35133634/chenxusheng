#### 1.简单柱形图

- 数据预处理

```python
course_count = course_orders.groupby("商品名称")["订单编号"].count().sort_values(ascending=False)
```

```python
x = course_count.index.tolist()
y = course_count.values.tolist()
```

- 直接生成柱形图

```python
from pyecharts import Bar

bar = Bar("松鼠学堂课程销量统计","付费课程学员数量统计")
bar.add("学员人数",x,y)
bar.render()
bar
```

- 对该柱形图进行美化

```python
from pyecharts import Bar,Style

style = Style(
    title_pos = "center",
    width=980,
    height=600,
    background_color="#fff"
)

style_bar = style.add(
    legend_top="bottom",        # 图例位置
    yaxis_label_textsize=9,     # y轴标签文字大小
    yaxis_rotate=45,            # y轴标签选择角度
    is_label_show = True,       # 展示柱形图上面的数值标签
    label_pos="right",          # 柱形图上面的数值标签显示位置
    label_text_size=12,         # 数值标签文本大小
    label_color=['skyblue']     # 柱形图颜色
)

bar = Bar("松鼠学堂课程销量统计","付费课程学员数量统计",**style.init_style)
bar.add("学员人数",x,y,is_convert=True,**style_bar)
bar.render()
bar
```

#### 4.dataZoom 效果，坐标轴缩放效果

- 数据预处理

```python
import pandas as pd
course_orders = pd.read_excel("./data/course_orders.xlsx")
```

```python
course_orders.head()
```

```python
from datetime import datetime

def dateFormat(s):
    try:
        return datetime.strptime(s,'%Y-%m-%d %H:%M:%S').date()
    except:
        return datetime.strptime(s,'%Y/%m/%d %H:%M').date()
```

```python
print(dateFormat('2018/03/28 13:36'))
```

```python
print(dateFormat('2018-03-20 15:08:00'))
```

```python
course_orders["订单日期"] = course_orders["订单时间"].apply(lambda s:dateFormat(str(s)))
```

```python
course_orders.head()
```

```python
date_orders = course_orders.groupby("订单日期").count()["订单编号"]
```

```python
date = date_orders.index.tolist()
counts = date_orders.values.tolist()
```

```python
from pyecharts import Bar,Style

style = Style(
    title_pos = "center",
    width=980,
    height=600,
    background_color="#fff"
)

style_bar = style.add(
    is_legend_show= False,    # 关闭图例
    label_color = ["red"],    # 柱子的颜色
    is_datazoom_show=True,
    datazoom_type='slider'
)


bar = Bar("每日销量展示",**style.init_style)
bar.add("课程销量", date,counts,**style_bar)
bar.render()
bar
```
